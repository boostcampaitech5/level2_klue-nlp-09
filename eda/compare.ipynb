```python
import pandas as pd
```

```python
origin = pd.read_csv('./prediction/submission.csv')
pl = pd.read_csv('submission.csv')
```

```python
length = len(origin)
same = 0
dif_list = []
for i in range(len(origin)):

    if origin['pred_label'][i] == pl['pred_label'][i]:
        same += 1
    else:
        dif_list.append([origin["pred_label"][i],pl["pred_label"][i]])
dif = length - same
print(length, same, dif)
for i in dif_list:
    print(i)
```

```python
count = 0

for i in dif_list:
    if "org" in i[0] and "org" not in i[1] and i[1] != "no_relation":
        print(i)
        count += 1
    elif "per" in i[0] and "per" not in i[1] and i[1] != "no_relation":
        print(i)
        count += 1
print(count)
```
